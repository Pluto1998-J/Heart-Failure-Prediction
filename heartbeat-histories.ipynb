```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("/kaggle/input/heart-failure-prediction-clinical-records/heart_failure_clinical_records.csv")
```

```python
data.head()
```

```python
plt.hist(data['age'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

```python
sns.countplot(data['anaemia'])
plt.xlabel('Anaemia (0: No, 1: Yes)')
plt.ylabel('Count')
plt.title('Count of Anaemia')
plt.show()
```

```python
sns.boxplot(data['creatinine_phosphokinase'])
plt.xlabel('CPK Levels')
plt.title('Box Plot of CPK Levels')
plt.show()
```

```python
sns.countplot(data['diabetes'])
plt.xlabel('Diabetes (0: No, 1: Yes)')
plt.ylabel('Count')
plt.title('Count of Diabetes')
plt.show()
```

```python
sns.boxplot(data['ejection_fraction'])
plt.xlabel('Ejection Fraction')
plt.title('Box Plot of Ejection Fraction')
plt.show()
```

```python
sns.countplot(data['high_blood_pressure'])
plt.xlabel('High Blood Pressure (0: No, 1: Yes)')
plt.ylabel('Count')
plt.title('Count of High Blood Pressure')
plt.show()
```

```python
sns.histplot(data['platelets'], kde=True, color='green')
plt.xlabel('Platelets')
plt.ylabel('Density')
plt.title('Distribution of Platelets')
plt.show()
```

```python
sns.histplot(data['serum_creatinine'], kde=True, color='orange')
plt.xlabel('Serum Creatinine')
plt.ylabel('Density')
plt.title('Distribution of Serum Creatinine')
plt.show()
```

```python
sns.boxplot(data['serum_sodium'])
plt.xlabel('Serum Sodium')
plt.title('Box Plot of Serum Sodium')
plt.show()
```

```python
sns.countplot(data['sex'])
plt.xlabel('Sex (0: Female, 1: Male)')
plt.ylabel('Count')
plt.title('Count of Gender')
plt.show()
```
